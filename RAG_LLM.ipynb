```python
import pypdf
filename ="/content/Brief_11544_217202393744K15.TTPT_NguyncHuy.pdf"
pdf_file = open(filename,'rb')

reader = pypdf.PdfReader(pdf_file)

content = ""
for page in reader.pages:
  text = page.extract_text()
  if text:
    content += text + "\n"
pdf_file.close()
```
